```python
import pygame
import math
from datetime import datetime

pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Clock")

clock_img = pygame.image.load("clock.png")
minute_hand = pygame.image.load("min_hand.png")
second_hand = pygame.image.load("sec_hand.png")

rrect = clock_img.get_rect(center=(400,300))
minute_rect = minute_hand.get_rect(center=(400,300))
second_rect = second_hand.get_rect(center=(400,300))

def blit1(img, angle, center):
    rot_img = pygame.transform.rotate(img, angle)
    rectt = rot_img.get_rect(center=center)
    screen.blit(rot_img, rectt.topleft)
def get_ang():
    now = datetime.now()
    sec_ang = -now.second * 6  
    min_ang = -now.minute * 6 - now.second * 0.1
    return sec_ang, min_ang

ff = True
while ff:
    screen.fill((255, 255, 255))
    screen.blit(clock_img, rrect.topleft)
    
    sec_ang, min_ang = get_ang()
    
    blit1(second_hand, sec_ang, rrect.center)
    blit1(minute_hand, min_ang, rrect.center)
    
    pygame.display.flip()
    pygame.time.delay(1000)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            ff = False
            
pygame.quit()
```

```python
import pygame
from pygame.locals import *
pygame.init()



def buttons():
    for label, (x, y, key) in buttons1.items():
        pygame.draw.rect(screen, (200, 200, 200), (x, y, button_w, button_h))
        font = pygame.font.Font(None, 24)
        text = font.render(label, True, (0, 0, 0))
        screen.blit(text, (x + 20, y + 10))




def play_music():
    pygame.mixer.music.play()

def stop_music():
    pygame.mixer.music.stop()

def next_song():
    global cur_song
    cur_song = (cur_song + 1) % len(playlist)
    pygame.mixer.music.load(playlist[cur_song])
    play_music()

def prev_song():
    global cur_song
    cur_song = (cur_song - 1) % len(playlist)
    pygame.mixer.music.load(playlist[cur_song])
    play_music()
 
screen = pygame.display.set_mode((500, 300))
pygame.display.set_caption("Musi—Å")
playlist = ["Radiohead - Creep.mp3", "NBSPLV - Downpour.mp3", "Blue Foundation - Grand.mp3"]
cur_song = 0
pygame.mixer.init()
pygame.mixer.music.load(playlist[cur_song])

button_w, button_h = 80, 40
buttons1 = {
    "Play": (100, 200, "P"),
    "Stop": (200, 200, "S"),
    "Next": (300, 200, "N"),
    "Prev": (400, 200, "B"),
}

running = True
while running:
    screen.fill((255, 255, 255))
    buttons()
    
    for event in pygame.event.get():
        if event.type == QUIT:
            running = False
        
        elif event.type == KEYDOWN:
            if event.key == K_p:  # Play
                play_music()
            elif event.key == K_s:  # Stop
                stop_music()
            elif event.key == K_n:  # Next song
                next_song()
            elif event.key == K_b:  # Prev song
                prev_song()
        
        elif event.type == MOUSEBUTTONDOWN:
            x, y = event.pos
            for label, (bx, by, key) in buttons.items():
                if bx <= x <= bx + button_w and by <= y <= by + button_h:
                    if key == "P":
                        play_music()
                    elif key == "S":
                        stop_music()
                    elif key == "N":
                        next_song()
                    elif key == "B":
                        prev_song()
    
    pygame.display.flip()

pygame.quit()

```

```python
import pygame
from pygame.locals import *
pygame.init()

screen = pygame.display.set_mode((500, 500))
pygame.display.set_caption("Balls")

r = 25
x, y =225, 225
speed = 20
clock = pygame.time.Clock()
running = True
while running:
    screen.fill((255, 255, 255))
    pygame.draw.circle(screen, (255, 0, 0), (x, y), r)
    pygame.display.flip()
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == QUIT:
            running = False
        elif event.type == KEYDOWN:
            if event.key == K_RIGHT and x + r + speed <= 500:
                x += speed
            elif event.key == K_LEFT and x - r - speed >= 0:
                x -= speed
            elif event.key == K_DOWN and y + r + speed <= 500:
                y += speed
            elif event.key == K_UP and y - r - speed >= 0:
                y -= speed

pygame.quit()
```
